```python
#create the GeoDataFrame
geometry = [Point(xy) for xy in zip(collision_uk_2023['longitude'], collision_uk_2023['latitude'])]
collision_gdf = gpd.GeoDataFrame(collision_uk_2023, geometry=geometry)
# 设置坐标系为 WGS84 （EPSG：4326）
collision_gdf.set_crs(epsg=4326, inplace=True)

#transfrom to GB
collision_gdf= collision_gdf.to_crs(boroughs.crs)

#execute space intersection
collision_london_intersect = gpd.sjoin(collision_gdf, boroughs, how='inner', predicate='intersects')
```

```python
#filter out rows(-1): the data missing or out of range
conditions = (
    (collision_gdf["weather_conditions"] != -1) &
    (collision_gdf["road_type"] != -1) &
    (collision_gdf["road_surface_conditions"] != -1) &
    (collision_gdf["speed_limit"] != -1)
)
collision_gdf_cleaned = collision_gdf[conditions]

#save the cleaned data to data/clean
collision_cleaned = collision_gdf_cleaned.to_csv(os.path.join("data/clean", "collision_cleaned.csv"),
                                                 index=False)
```

- normalization: **z-score algorithm**\
  The method is based on the mean value and standard deviation of the data for normalization processing, the steps are as follows:
  1) to obtain the arithmetic mean value μ and standard deviation s of each characteristic value of the data to be processed respectively.
  2) Carry out standardization processing according to formula (below), and find the standardized value of each characteristic respectively.\
  z = (x - μ) /s

```python
from scipy.stats import zscore

# 对指定列进行 Z 分数标准化
columns_to_standardize = ['weather_conditions',
                          'day_of_week',
                          'road_type',
                          'road_surface_conditions',
                          'speed_limit',
                          'number_of_casualties']
#create a new dataframe
collision_gdf_standardized = collision_gdf_cleaned.copy()

# 对每个指定的列进行 Z 分数标准化
for column in columns_to_standardize:
    # 确保列存在且为数值类型
    if column in collision_gdf_cleaned.columns and collision_gdf_cleaned[column].dtype in ['int64', 'float64']:
        collision_gdf_standardized[column + '_zscore'] = zscore(collision_gdf_cleaned[column])
    else:
        print(f"列 {column} 不存在或不是数值类型，跳过标准化。")

# 打印结果
print(collision_gdf_standardized.head())
```

- 因子检测

```python
# 执行因子分析
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(n_factors=2, rotation="varimax")
fa.fit(subset)

# 获取因子载荷矩阵
factor_loadings = fa.loadings_
print("\n因子载荷矩阵:")
print(pd.DataFrame(factor_loadings, index=subset.columns))

# 获取因子得分系数矩阵
factor_score_coeff = fa.transform(subset)
print("\n因子得分系数矩阵:")
print(pd.DataFrame(factor_score_coeff, columns=["Factor1", "Factor2"]))
```

```python
# 使用肘部法则确定聚类数量
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def determine_optimal_clusters(data, max_clusters=10):
    inertia = []
    silhouette_scores = []
    cluster_range = range(2, max_clusters + 1)

    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)  # SSE (Sum of Squared Errors)
        silhouette_scores.append(silhouette_score(data, kmeans.labels_))

    # 绘制肘部法则图
    plt.figure(figsize=(10, 5))
    plt.plot(cluster_range, inertia, marker='o', label='Inertia (SSE)')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia (SSE)')
    plt.title('Elbow Method for Optimal Clusters')
    plt.legend()
    plt.show()

    # 绘制轮廓系数图
    plt.figure(figsize=(10, 5))
    plt.plot(cluster_range, silhouette_scores, marker='o', label='Silhouette Score')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Method for Optimal Clusters')
    plt.legend()
    plt.show()

    return inertia, silhouette_scores

# 决定最佳聚类数
factor_score = pd.DataFrame(factor_score_coeff, columns=["Factor1", "Factor2"])
determine_optimal_clusters(factor_score)
```

```python
# 使用 KMeans 聚类
optimal_clusters = 3  # 根据上面的图选择合适的 k 值
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
factor_score['Cluster'] = kmeans.fit_predict(factor_score)

# 输出聚类结果
print(factor_score.head())

# 可视化聚类结果
plt.figure(figsize=(8, 6))
plt.scatter(factor_score['Factor1'], factor_score['Factor2'], c=factor_score['Cluster'], cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.xlabel('Factor1')
plt.ylabel('Factor2')
plt.title('K-Means Clustering of Factor Scores')
plt.legend()
plt.show()
```
